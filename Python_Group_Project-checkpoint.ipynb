```python
#importing the important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading and concatenating the data
df0 = pd.read_csv("2000-01.csv")
df1 = pd.read_csv("2001-02.csv")
df2 = pd.read_csv("2002-03.csv")
df3 = pd.read_csv("2003-04.csv")
df4 = pd.read_csv("2004-05.csv")
df5 = pd.read_csv("2005-06.csv")
df6 = pd.read_csv("2006-07.csv")
df7 = pd.read_csv("2007-08.csv")
df8 = pd.read_csv("2008-09.csv")
df9 = pd.read_csv("2009-10.csv")
df10 = pd.read_csv("2010-11.csv")
df11 = pd.read_csv("2011-12.csv")
df12 = pd.read_csv("2012-13.csv")
df13 = pd.read_csv("2013-14.csv")
df14 = pd.read_csv("2014-15.csv")
df15 = pd.read_csv("2015-16.csv")
df16 = pd.read_csv("2016-17.csv")
df17 = pd.read_csv("2017-18.csv")
df18 = pd.read_csv("2018-19.csv")
df19 = pd.read_csv("2019-20.csv")
```

## Visualization

### 1. Season Wise Total Goals Scored

```python
#combining the away goals and home goals scored columns
main_data['Total Goals'] = main_data['FTHG'] + main_data['FTAG']
    
# Group by year and sum total goals
yearly_goals = main_data.groupby('Year')['Total Goals'].sum().reset_index()
    
# Rename columns for clarity
yearly_goals.columns = ['Year', 'Total Goals']
    
# Plot the bar graph
plt.figure(figsize=(12, 6))
plt.bar(yearly_goals['Year'], yearly_goals['Total Goals'], color='skyblue', edgecolor='black')
plt.title('Total Goals Scored Each Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Goals', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
    
# Show the plot
plt.tight_layout()
plt.show()
```

### Home vs Away Goals Per Year

```python
#calculating home and away goals separately
yearly_home_goals = main_data.groupby('Year')['FTHG'].sum().reset_index()
yearly_away_goals = main_data.groupby('Year')['FTAG'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.bar(yearly_home_goals['Year'] - 0.2, yearly_home_goals['FTHG'], width=0.4, label='Home Goals', color='blue')
plt.bar(yearly_away_goals['Year'] + 0.2, yearly_away_goals['FTAG'], width=0.4, label='Away Goals', color='orange')
plt.title('Home vs Away Goals Per Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Goals', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

###  Average Goals Per Match

```python
#calculating the average of all goals
main_data['Total Goals'] = main_data['FTHG'] + main_data['FTAG']
average_goals = main_data.groupby('Year')['Total Goals'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(average_goals['Year'], average_goals['Total Goals'], marker='o', color='green', label='Average Goals')
plt.title('Average Goals Per Match Per Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Goals', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

###  Top-Scoring Matches

```python
main_data['Total Goals'] = main_data['FTHG'] + main_data['FTAG']

top_matches = main_data.nlargest(10, 'Total Goals')

top_matches['Match'] = (
    top_matches['Year'].astype(str) + ": " +
    top_matches['HomeTeam'] + " vs " + 
    top_matches['AwayTeam']
)

plt.figure(figsize=(12, 8))
plt.barh(top_matches['Match'], top_matches['Total Goals'], color='purple', edgecolor='black')

plt.title('Top 10 Highest Scoring Matches', fontsize=16)
plt.xlabel('Total Goals', fontsize=14)
plt.ylabel('Match', fontsize=14)
plt.gca().invert_yaxis()  
plt.tight_layout()

plt.show()
```

### Home Percentage win vs Away Percentage Win

```python
home_wins = (main_data['FTHG'] > main_data['FTAG']).sum()  # Home wins
away_wins = (main_data['FTHG'] < main_data['FTAG']).sum()  # Away wins
draws = (main_data['FTHG'] == main_data['FTAG']).sum()     # Draws

# Total number of matches
total_matches = home_wins + away_wins + draws

# Percentages
home_win_percentage = (home_wins / total_matches) * 100
away_win_percentage = (away_wins / total_matches) * 100
draw_percentage = (draws / total_matches) * 100

# Data for the pie chart
labels = ['Home Wins', 'Away Wins', 'Draws']
percentages = [home_win_percentage, away_win_percentage, draw_percentage]
colors = ['#ff9999', '#66b3ff', '#99ff99']

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=labels, colors=colors, autopct='%.1f%%', startangle=90, textprops={'fontsize': 12})
plt.title('Match Outcomes: Home Wins, Away Wins, and Draws', fontsize=16)
plt.show()
```

## Changes in Goal Scoring Over the Years:
## Increasing or decreasing trends in goals scored over seasons can indicate shifts in playing styles or competitiveness. Sponsors might align with leagues where scoring is increasing.

```python


# Define close matches:
# Matches decided by one goal
close_matches = main_data[(abs(main_data['FTHG'] - main_data['FTAG']) == 1)]

# High-scoring draws
high_scoring_draws = main_data[(main_data['FTHG'] == main_data['FTAG']) & (main_data['FTHG'] > 2)]

# Count close matches and high-scoring draws
close_match_count = len(close_matches)
high_scoring_draw_count = len(high_scoring_draws)

# Total matches
total_matches = len(main_data)

# Prepare data for visualization
labels = ['Close Matches (1 Goal Margin)', 'High-Scoring Draws (3+ Goals)', 'Other Matches']
counts = [close_match_count, high_scoring_draw_count, total_matches - (close_match_count + high_scoring_draw_count)]
colors = ['#ff9999', '#66b3ff', '#99ff99']

# Plotting the Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=labels, colors=colors, autopct='%.1f%%', startangle=90, textprops={'fontsize': 12})
plt.title('Match Tension Analysis: Close Matches and High-Scoring Draws', fontsize=16)
plt.show()

# Analyzing close matches further (Top rivalries in close matches)
if 'HomeTeam' in main_data.columns and 'AwayTeam' in main_data.columns:
    rivalries = close_matches.groupby(['HomeTeam', 'AwayTeam']).size().reset_index(name='Count')
    top_rivalries = rivalries.nlargest(10, 'Count')

    # Bar Plot for Top Rivalries
    plt.figure(figsize=(12, 6))
    plt.barh(top_rivalries.apply(lambda x: f"{x['HomeTeam']} vs {x['AwayTeam']}", axis=1), 
             top_rivalries['Count'], color='purple', edgecolor='black')
    plt.title('Top 10 Rivalries in Close Matches', fontsize=16)
    plt.xlabel('Number of Close Matches', fontsize=14)
    plt.ylabel('Rivalries', fontsize=14)
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```

```python


# Filter for home wins
home_wins = main_data[main_data['FTHG'] > main_data['FTAG']]

# Count home wins for each team
home_win_counts = home_wins['HomeTeam'].value_counts()

# Count total home games for each team
total_home_games = main_data['HomeTeam'].value_counts()

# Calculate home win percentage for each team
home_win_percentage = (home_win_counts / total_home_games) * 100

# Sort teams by home win percentage
home_win_percentage = home_win_percentage.sort_values(ascending=False)

# Select top teams for visualization
top_teams = home_win_percentage.head(10)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_teams.index, top_teams.values, color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Top 10 Teams with Highest Home Win Percentage', fontsize=16)
plt.xlabel('Teams', fontsize=14)
plt.ylabel('Home Win Percentage (%)', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()

print("If certain teams dominate at home consistently, this insight is valuable for broadcasters to create narratives around fortress-like stadiums.")
```

```python
def is_come_from_behind_win(row):

    if row["FTR"] == "H" and row["HTHG"] < row["HTAG"]:

        return True

    elif row["FTR"] == "A" and row["HTAG"] < row["HTHG"]:

        return True

    return False

# Filter the dataset

main_data["ComebackWin"] = main_data.apply(is_come_from_behind_win, axis=1)

comeback_wins = main_data[main_data["ComebackWin"]]

# Display the results

print("Matches where a team was trailing at halftime but won:")

print(comeback_wins[["HomeTeam", "AwayTeam", "HTHG", "HTAG", "FTR"]])
```

### Shot Accuracy
Insight: Comparing shots on target to total shots gives the efficiency of home and away teams in converting attempts into accurate shots.

What you can learn: Which teams (home or away) are more precise in their attacks, and how effective teams are at finishing their opportunities.

```python
# Shot accuracy calculation with division by zero handling
main_data['Home Accuracy'] = np.where(main_data['HS'] != 0, main_data['HST'] / main_data['HS'] * 100, 0)
main_data['Away Accuracy'] = np.where(main_data['AS'] != 0, main_data['AST'] / main_data['AS'] * 100, 0)

print("Shot Accuracy Statistics:")
print(f"Home Team Accuracy: {main_data['Home Accuracy'].mean():.2f}%")
print(f"Away Team Accuracy: {main_data['Away Accuracy'].mean():.2f}%")

```

### Referee Bias (Yellow/Red Cards)
Insight: Checking which referees issue the most cards can highlight potential biases or stricter officiating styles.

What you can learn: Do some referees have reputations for strictness? This might affect team strategies or fan perceptions.

```python
# Referee bias analysis
referee_cards = main_data.groupby('Referee')[['HY', 'AY', 'HR', 'AR']].sum()
referee_cards['Total Yellow Cards'] = referee_cards['HY'] + referee_cards['AY']
referee_cards['Total Red Cards'] = referee_cards['HR'] + referee_cards['AR']

print("Referee Bias - Average Cards Given:")
print(referee_cards.sort_values(by='Total Yellow Cards', ascending=False).head())
```

###  Card Aggression
Insight: Examining the yellow and red cards received by teams indicates their aggression levels and disciplinary records.

What you can learn: Which teams play more aggressively and whether this impacts their performance positively or negatively.

```python
# Sum yellow and red cards for home and away teams separately
home_cards = main_data.groupby('HomeTeam')[['HY', 'HR']].sum()
away_cards = main_data.groupby('AwayTeam')[['AY', 'AR']].sum()

# Rename columns for away team cards
away_cards.columns = ['Away Yellow Cards', 'Away Red Cards']

# Combine home and away cards data
team_cards = home_cards.add(away_cards, fill_value=0)

# Calculate total cards
team_cards['Total Yellow Cards'] = team_cards['HY'] + team_cards['Away Yellow Cards']
team_cards['Total Red Cards'] = team_cards['HR'] + team_cards['Away Red Cards']

# Rename columns for clarity
team_cards.columns = ['Home Yellow Cards', 'Home Red Cards', 'Away Yellow Cards', 'Away Red Cards', 'Total Yellow Cards', 'Total Red Cards']

print("Teams with Most Cards:")
print(team_cards.sort_values(by='Total Yellow Cards', ascending=False).head())
```

```python

# Plotting the yellow and red cards analysis
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot for yellow cards
team_cards[['Total Yellow Cards']].sort_values(by='Total Yellow Cards', ascending=False).head(10).plot(kind='barh', ax=ax[0], color='#FFD700', edgecolor='black')
ax[0].set_title('Teams with Most Yellow Cards')
ax[0].set_xlabel('Total Yellow Cards')
ax[0].set_ylabel('Teams')

# Plot for red cards
team_cards[['Total Red Cards']].sort_values(by='Total Red Cards', ascending=False).head(10).plot(kind='barh', ax=ax[1], color='#FF4500', edgecolor='black')
ax[1].set_title('Teams with Most Red Cards')
ax[1].set_xlabel('Total Red Cards')
ax[1].set_ylabel('Teams')

plt.tight_layout()
plt.show()

```

### Most Dominant Teams at Home and Away
What it means: Teams with the highest win rates at home and away reflect dominance in their respective conditions. High home win rates may indicate strong fan support or familiarity with the home pitch, while high away win rates showcase resilience and adaptability.

What you can learn: Which teams are nearly unbeatable at home and which ones perform well under challenging away conditions.

```python
# Home win rate
home_wins = main_data[main_data['FTR'] == 'H'].groupby('HomeTeam').size()
home_matches = main_data.groupby('HomeTeam').size()
home_win_rate = (home_wins / home_matches).sort_values(ascending=False)

# Away win rate
away_wins = main_data[main_data['FTR'] == 'A'].groupby('AwayTeam').size()
away_matches = main_data.groupby('AwayTeam').size()
away_win_rate = (away_wins / away_matches).sort_values(ascending=False)

print("Top Home Teams by Win Rate:")
print(home_win_rate.head())
print("\nTop Away Teams by Win Rate:")
print(away_win_rate.head())
```

### Team Discipline
Find the most disciplined and least disciplined teams (based on cards received).

```python
# Sum yellow and red cards for home and away teams separately
home_discipline = main_data.groupby('HomeTeam')[['HY', 'HR']].sum()
away_discipline = main_data.groupby('AwayTeam')[['AY', 'AR']].sum()

# Rename columns for away team discipline
away_discipline.columns = ['Away Yellow Cards', 'Away Red Cards']

# Combine home and away discipline data
discipline = home_discipline.add(away_discipline, fill_value=0)

# Calculate total cards
discipline['Total Cards'] = discipline['HY'] + discipline['HR'] + discipline['Away Yellow Cards'] + discipline['Away Red Cards']

# Rename columns for clarity
discipline.columns = ['Home Yellow Cards', 'Home Red Cards', 'Away Yellow Cards', 'Away Red Cards', 'Total Cards']

print("Most Disciplined Teams:")
print(discipline.sort_values(by='Total Cards').head())

print("\nLeast Disciplined Teams:")
print(discipline.sort_values(by='Total Cards', ascending=False).head())

```

```python
# Plotting the discipline analysis
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

# Plot for most disciplined teams
discipline.sort_values(by='Total Cards').head(10).plot(kind='barh', ax=ax[0], color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], edgecolor='black')
ax[0].set_title('Most Disciplined Teams (Least Total Cards)')
ax[0].set_xlabel('Total Cards')
ax[0].set_ylabel('Teams')

# Plot for least disciplined teams
discipline.sort_values(by='Total Cards', ascending=False).head(10).plot(kind='barh', ax=ax[1], color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], edgecolor='black')
ax[1].set_title('Least Disciplined Teams (Most Total Cards)')
ax[1].set_xlabel('Total Cards')
ax[1].set_ylabel('Teams')

plt.tight_layout()
plt.show()
```

### Effectiveness of Shots (Conversion Rate)
What it means: The conversion rate measures how efficiently teams convert shots into goals. High conversion rates indicate sharp attacking play, while low rates suggest poor finishing or excellent opposition defense.

What you can learn: Which teams capitalize on their chances the best and which ones struggle to turn opportunities into results.

```python
# Conversion rate
import numpy as np

# Conversion rate calculation with division by zero handling
main_data['Home Conversion Rate'] = np.where(main_data['HS'] != 0, main_data['FTHG'] / main_data['HS'] * 100, 0)
main_data['Away Conversion Rate'] = np.where(main_data['AS'] != 0, main_data['FTAG'] / main_data['AS'] * 100, 0)

avg_home_conversion = main_data['Home Conversion Rate'].mean()
avg_away_conversion = main_data['Away Conversion Rate'].mean()

print(f"Average Home Conversion Rate: {avg_home_conversion:.2f}%")
print(f"Average Away Conversion Rate: {avg_away_conversion:.2f}%")

```

### Best Defensive Teams
What it means: Teams that concede the fewest goals at home and away are usually strong defensively. This insight highlights their ability to neutralize opposition attacks.

What you can learn: Which teams have solid defenses, making them hard to break down regardless of the opponent or location.

```python
# Defensive performance
home_conceded = main_data.groupby('HomeTeam')['FTAG'].sum() / home_matches
away_conceded = main_data.groupby('AwayTeam')['FTHG'].sum() / away_matches

print("Best Home Defensive Teams (Least Goals Conceded):")
print(home_conceded.sort_values().head())

print("\nBest Away Defensive Teams (Least Goals Conceded):")
print(away_conceded.sort_values().head())

```

```python

# Plotting the defensive performance
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot for home defensive performance
home_conceded.sort_values().head(10).plot(kind='barh', ax=ax[0], color='skyblue', edgecolor='black')
ax[0].set_title('Best Home Defensive Teams (Least Goals Conceded)')
ax[0].set_xlabel('Goals Conceded')
ax[0].set_ylabel('Teams')

# Plot for away defensive performance
away_conceded.sort_values().head(10).plot(kind='barh', ax=ax[1], color='lightcoral', edgecolor='black')
ax[1].set_title('Best Away Defensive Teams (Least Goals Conceded)')
ax[1].set_xlabel('Goals Conceded')
ax[1].set_ylabel('Teams')

plt.tight_layout()
plt.show()
```
